```python
# df.rename(mapper={'make':'brand'})   # default search takes plaace row wise so it wont change anything 
df.rename(mapper={"make":"brand"}, axis=1 , inplace= True)  # inplace returs the actual change in data 
#df.rename(mapper={"make":"brand","previous name":"new name"}, axis=1 , inplace= True)
```

```python
df.head()
```

```python
# how to drop a column 
df.drop(columns=['Unnamed: 0'],inplace= True)
```

```python
df.head()
```

```python
df.replace('alfa-romero','alfa-romeo',inplace=True)
```

```python
df.head()
```

```python
df.replace('?',np.nan,inplace=True)  # removes all ? by np.nan = not a number
```

```python
df.head()
```

```python
df.dropna(subset=['price'],inplace=True)     # droped some records where the value of price was missing 
```

```python
df.head()
```

```python
df.isnull().sum()    # conut the number of missing data in each column
```

```python
# changing missing values with the mean of a column 

mnl = df['normalized-losses'].astype('float').mean()
```

```python
df['normalized-losses'].replace(np.nan,mnl,inplace=True)
```

```python
df.head()
```

```python
mnl1 = df['bore'].astype('float').mean()
```

```python
df['bore'].replace(np.nan,mnl1,inplace=True)
```

```python
mnl2 = df['stroke'].astype('float').mean()
```

```python
df['stroke'].replace(np.nan,mnl2,inplace=True)
df['horsepower'].replace(np.nan,df['horsepower'].astype(float).mean(),inplace=True)
df['peak-rpm'].replace(np.nan,df['peak-rpm'].astype(float).mean(),inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].replace(np.nan,'four',inplace=True)
```

```python
df.isnull().sum()
```

```python
df.to_csv("automobile_cleaned.csv",index= False)
```
