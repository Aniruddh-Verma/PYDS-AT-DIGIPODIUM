```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [21,345,56,7934,382,45,35]
plt.plot(y)
plt.show()
```

```python
plt.figure(figsize = (15,4))
plt.pie(y)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize= (15,5))
plt.show()
```

```python
t = np.arange(-3*np.pi, 3*np.pi, 0.2)
x = np.sin(t)
y = np.cos(t)
fig , ax = plt.subplots(nrows = 1, ncols = 2, figsize= (15,2))
ax[0].plot(t,x,'ro')
ax[1].plot(t,y,'b')
plt.show()
```

```python
x = [100,200,300,400,500,600,700,800,900,1000]
y = [23,45,1,23,45,34,51,14,45,56]
plt.plot(x, y, linewidth=2, color='blue', marker='o', markersize=15, markerfacecolor='red', linestyle='-')
#plt.plot(x, y,'go--', linewidth=2, markersize=15, markerfacecolor='red')
plt.show()
```

```python
plt.plot(x, y, linewidth=2, color='blue', marker='o', markersize=15, markerfacecolor='red', linestyle='-')
plt.title('Some Graph Showing Random Data', fontdict={'fontsize':15,'fontfamily':'Arial', 'fontweight':'light','color':'green'})
plt.xlabel('X axis ki Values')
plt.ylabel('Y axis ki values',fontdict={'color':'red'})
plt.text(800,5, 'This is Serious\ntext here')
plt.annotate("lowest value", xy = (300,1),xytext=(500,20),arrowprops=dict(facecolor ='black', shrink = 0.05))
plt.show()
```
