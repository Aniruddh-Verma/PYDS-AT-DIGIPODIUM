### 3.BREAK AND CONTINUE

##### 1.BREAK STATEMENT 

```python
# EXAMPLE 1: BREAK STATEMENT

x=[1,2,3,4,0,0,5]
for i in x:
      if i==0:
            break
      print(i)
```

##### 2. CONTINUE STATEMENT 

```python
# EXAMPLE 2: CONTINUE STATEMENT
x=[1,2,3,4,0,0,5]
for i in x:
      if i==0:
            continue
      print(i)

## only use for memory optimisation
```

```python
# METHOD 1: WAP TO CHECK WHETHER THE NUMBER IS PRIME NUMBER 
x=int(input("enter the number"))
count=0
for i in range(1,x+1):
      if x%i == 0:
            count+=1      
if count == 2:
      print('prime number')
elif count !=2:
      print('not prime')
```

```python
# METHOD 2: WAP TO CHECK WHETHER THE NUMBER IS PRIME NUMBER 
num=11
for i in range(2,num):
      if num %i ==0:
            print('not prime')
            break
else:   #else block is only execute when for is broken
      print('prime')
```

### 4. WHILE LOOP 

```python
# EXAMPLE 1:

x=10
while(x>0):
      print(x)
      x-=2
```

```python
# EXAMPLE 2:
total=0
while True:
      num= int(input('enter a number'))
      if num == 0:
            break
      total+= num
print(f'total is {total}')
```

```python
#EXAMPLE 3:
book=[]
while True:
      name= input('enter the name: ')
      if name:
            book.append(name)
      else:
            break
print('these are your books')
print(book)
```

```python
# WAP TO PRINT THE FIBBONACCI SERIES UPTO TO THE GIVEN NUMBER

num= 10
n1,n2=0,1
print('fibbonacci series:',n1,n2,end=',')
for i in range(2,num):
      n3=n1+n2
      print(n3,end=',')
      n1=n2
      n2=n3
print()
```
